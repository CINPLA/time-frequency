Make a chirp

```python
%pylab inline
rcParams['figure.figsize'] = (15, 9)
import pywt
```

```python
duration = 10
sampling_rate = 1000
sampling_period = 1 / sampling_rate
time = arange(0, duration, sampling_period)
f_start = 0
f_stop = 500
amplitude = 10

frequency = f_start + (f_stop - f_start) * time / (2 * duration)
data = amplitude * np.sin(2 * np.pi * frequency * time)

fig, axs = subplots(2, 1, sharex=True)
axs[0].plot(time, data)

axs[1].specgram(data, Fs=sampling_rate);
```

```python
pywt.wavelist(kind='continuous')
```

```python
wavelet = 'cmor1.5-1.0'
scales = np.arange(1, 100, 2)

power, freq = pywt.cwt(data, scales, wavelet, sampling_period)
# power = (abs(cfs)) ** 2
pywt.WaveletPacket(data, 'db2', 'sym', maxlevel=5)
imshow(abs(power), extent=[0, duration, freq.min(), freq.max()], aspect='auto')
```

```python
duration = 10
sampling_rate = 1000
time = arange(0, duration, 1/sampling_rate)
f_start = 1
f_stop = 500
amplitude = 10

k = (f_stop / f_start) **(1 / duration)

signal = amplitude * sin(2 * np.pi * f_start * (k**time - 1) / log(k))

fig, axs = subplots(2, 1, sharex=True)
axs[0].plot(time, signal)

axs[1].specgram(signal, Fs=sampling_rate);
```
